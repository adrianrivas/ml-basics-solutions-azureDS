The dataset contains observations of US domestic flights in 2013, and consists of the following fields:

- **Year**: The year of the flight (all records are from 2013)
- **Month**: The month of the flight
- **DayofMonth**: The day of the month on which the flight departed
- **DayOfWeek**: The day of the week on which the flight departed - from 1 (Monday) to 7 (Sunday)
- **Carrier**: The two-letter abbreviation for the airline.
- **OriginAirportID**: A unique numeric identifier for the departure aiport
- **OriginAirportName**: The full name of the departure airport
- **OriginCity**: The departure airport city
- **OriginState**: The departure airport state
- **DestAirportID**: A unique numeric identifier for the destination aiport
- **DestAirportName**: The full name of the destination airport
- **DestCity**: The destination airport city
- **DestState**: The destination airport state
- **CRSDepTime**: The scheduled departure time
- **DepDelay**: The number of minutes departure was delayed (flight that left ahead of schedule have a negative value)
- **DelDelay15**: A binary indicator that departure was delayed by more than 15 minutes (and therefore considered "late")
- **CRSArrTime**: The scheduled arrival time
- **ArrDelay**: The number of minutes arrival was delayed (flight that arrived ahead of schedule have a negative value)
- **ArrDelay15**: A binary indicator that arrival was delayed by more than 15 minutes (and therefore considered "late")
- **Cancelled**: A binary indicator that the flight was cancelled

#### Outliers Detection

```python
def show_distribution(var_data):
    from matplotlib import pyplot as plt

    # Get statistics
    min_val = var_data.min()
    max_val = var_data.max()
    mean_val = var_data.mean()
    med_val = var_data.median()
    mod_val = var_data.mode()[0]

    print('Minimum:{:.2f}\nMean:{:.2f}\nMedian:{:.2f}\nMode:{:.2f}\nMaximum:{:.2f}\n'.format(min_val,
                                                                                            mean_val,
                                                                                            med_val,
                                                                                            mod_val,
                                                                                            max_val))

    # Create a figure for 2 subplots (2 rows, 1 column)
    fig, ax = plt.subplots(2, 1, figsize = (10,4))

    # Plot the histogram   
    ax[0].hist(var_data)
    ax[0].set_ylabel('Frequency')

    # Add lines for the mean, median, and mode
    ax[0].axvline(x=min_val, color = 'gray', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=mean_val, color = 'cyan', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=med_val, color = 'red', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=mod_val, color = 'yellow', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=max_val, color = 'gray', linestyle='dashed', linewidth = 2)

    # Plot the boxplot   
    ax[1].boxplot(var_data, vert=False)
    ax[1].set_xlabel('Value')

    # Add a title to the Figure
    fig.suptitle('Data Distribution')

    # Show the figure
    fig.show()
```

```python
show_distribution(df_flights['DepDelay'])
```

```python
show_distribution(df_flights['ArrDelay'])
```

Both columns have a similar distribution, where many outliers are detected in the positive side of each column, which has some of logic, if we suppose that one flight would have the same delay time at its departure and arrival.

We are going to remove all the rows with values after 150 minutes.

```python
df_flights_not_outliers = df_flights[(df_flights['DepDelay'] <= 150) & (df_flights['ArrDelay'] <= 150)]
```

Let's veirfy the new distributions.

```python
show_distribution(df_flights_not_outliers['DepDelay'])
```

```python
show_distribution(df_flights_not_outliers['ArrDelay'])
```

Despite the outliers elimination, the box plot still is detecting outliers after -25 and 25 minutes, for that reason, we are going to do a second outliers removal.

```python
df_flights_not_outliers = df_flights_not_outliers[(df_flights_not_outliers['DepDelay'] <= 25) & \
                                                  (df_flights_not_outliers['DepDelay'] >= -25) & \
                                                  (df_flights_not_outliers['ArrDelay'] <= 25) & \
                                                  (df_flights_not_outliers['ArrDelay'] >= -25)
                                                 ]
```

```python
show_distribution(df_flights_not_outliers['DepDelay'])
```

```python
show_distribution(df_flights_not_outliers['ArrDelay'])
```

Now, we are going to graph the density function of both columns.

```python
def show_density(var_data):
    from matplotlib import pyplot as plt

    fig = plt.figure(figsize=(10,4))

    # Plot density
    var_data.plot.density() #Gráfico para obtener la densidad de probabilidad

    # Add titles and labels
    plt.title('Data Density')

    # Show the mean, median, and mode
    plt.axvline(x=var_data.mean(), color = 'cyan', linestyle='dashed', linewidth = 2)
    plt.axvline(x=var_data.median(), color = 'red', linestyle='dashed', linewidth = 2)
    plt.axvline(x=var_data.mode()[0], color = 'yellow', linestyle='dashed', linewidth = 2)

    # Show the figure
    plt.show()
```

```python
show_density(df_flights_not_outliers['DepDelay'])
```

```python
show_density(df_flights_not_outliers['ArrDelay'])
```

### 2. Explore the Cleaned Data

#### View summary statistics for the numeric fields in the dataset.

```python
df_flights_cleaned = df_flights_not_outliers.copy()
```

```python
df_flights_cleaned.describe()
```

We are going to change the data type of the column Year, OriginAirportID, DestAirportID, CRSDepTime	& CRSArrTime, because because even though these are numeric columns, we are not going any calculation with them.

Also, we are going to change the data type from integer to boolean for the columns DepDel15, ArrDel15 & Cancelled, because we want just to describe the real numeric columns.

```python
df_flights_cleaned['Year'] = df_flights_cleaned['Year'].astype(str)
df_flights_cleaned['OriginAirportID'] = df_flights_cleaned['OriginAirportID'].astype(str)
df_flights_cleaned['DestAirportID'] = df_flights_cleaned['DestAirportID'].astype(str)
df_flights_cleaned['CRSDepTime'] = df_flights_cleaned['CRSDepTime'].astype(str)
df_flights_cleaned['CRSArrTime'] = df_flights_cleaned['CRSArrTime'].astype(str)
df_flights_cleaned['DepDel15'] = df_flights_cleaned['DepDel15'].astype(bool)
df_flights_cleaned['ArrDel15'] = df_flights_cleaned['ArrDel15'].astype(bool)
df_flights_cleaned['Cancelled'] = df_flights_cleaned['Cancelled'].astype(bool)
```

We only leave the real numeric columns and see the summary statistics only for these columns.

```python
df_flights_cleaned.describe()
```

#### Determine the distribution of the DepDelay and ArrDelay columns.

```python
from matplotlib import pyplot as plt

fig, ax = plt.subplots(1, 2, figsize = (15,5))

df_flights_cleaned['DepDelay'].plot.density(ax=ax[0])
df_flights_cleaned['ArrDelay'].plot.density(ax=ax[1])

ax[0].set_title('DepDelay Distribution')
ax[1].set_title('ArrDelay Distribution')

fig.suptitle('Distributions')

fig.show()
```

#### What are the average (mean) departure and arrival delays?

```python
print('Average Departure Delays: {}'.format(round(df_flights_cleaned['DepDelay'].mean(), 2)))
print('Average Arrival Delays: {}'.format(round(df_flights_cleaned['ArrDelay'].mean(), 2)))
```

#### How do the carriers compare in terms of arrival delay performance?

```python
df_flights_cleaned
```

We only keep the rows where ArrDelay is higher than 0, in other words, flights that got some delay time in arrival.

```python
df_flights_grouped = df_flights_cleaned[df_flights_cleaned['ArrDelay'] > 0]
df_flights_grouped = df_flights_grouped.groupby('Carrier')['ArrDelay'].mean()
df_flights_grouped = df_flights_grouped.reset_index()
df_flights_grouped.rename({'ArrDelay': 'AvgArrDelay'}, axis=1, inplace=True)
df_flights_grouped
```

```python
df_flights_grouped.plot.bar(x='Carrier', y='AvgArrDelay', color='teal', figsize=(6,6))
```

In terms of arrival delay, the performance by carrier is kind of similar, but it's important to mention that on average:

- The top 3 carriers that have the highest arrival delay time are **MQ, EV & B6.**
- The carrier that has the lowest arrival delay time is **HA.**
- The rest of carriers arrival delay perfomances are too similar.

#### Is there a noticable difference in arrival delays for different days of the week?

```python
df_flights_grouped = df_flights_cleaned[df_flights_cleaned['ArrDelay'] > 0]
df_flights_grouped = df_flights_grouped.groupby('DayOfWeek')['ArrDelay'].mean()
df_flights_grouped = df_flights_grouped.reset_index()
df_flights_grouped.rename({'ArrDelay': 'AvgArrDelay'}, axis=1, inplace=True)
df_flights_grouped
```

```python
df_flights_grouped.plot.bar(x='DayOfWeek', y='AvgArrDelay', color='teal', figsize=(6,8))
```

The delay time in arrivals by day of week is also similar. Thus, there is not a noticable difference between one or another day.

#### Which departure airport has the highest average departure delay?

```python
df_flights_grouped = df_flights_cleaned[df_flights_cleaned['DepDelay'] > 0]
df_flights_grouped = df_flights_grouped.groupby('OriginAirportName')['DepDelay'].mean()
df_flights_grouped = df_flights_grouped.reset_index()
df_flights_grouped = df_flights_grouped.sort_values('DepDelay', ascending=False)
df_flights_grouped.index = range(df_flights_grouped.shape[0])
df_flights_grouped.index = df_flights_grouped.index + 1
df_flights_grouped.rename({'DepDelay': 'AvgDepDelay'}, axis=1, inplace=True)
df_flights_grouped
```

The airport departure that has the highest average departure delay is **Chicago Midway International**.

#### Do late departures tend to result in longer arrival delays than on-time departures?

- Late departures are going to be identified with the column DepDel15 == True
- On-time departures are going to be identified with the column DepDel15 == False

```python
df_flights_grouped = df_flights_cleaned[df_flights_cleaned['ArrDelay'] > 0]
df_flights_grouped['Late_Indicator'] = np.where(df_flights_grouped['DepDel15'] == True, 'Late Departure', 'On-time Departure')
df_flights_grouped = df_flights_grouped.groupby('Late_Indicator')['ArrDelay'].mean()
df_flights_grouped = df_flights_grouped.reset_index()
df_flights_grouped = df_flights_grouped.sort_values('ArrDelay', ascending=False)
df_flights_grouped.index = range(df_flights_grouped.shape[0])
df_flights_grouped.index = df_flights_grouped.index + 1
df_flights_grouped.rename({'ArrDelay': 'AvgArrDelay'}, axis=1, inplace=True)
df_flights_grouped
```

Late Departure has a higher average arrival delay than on-time departures.

#### Which route (from origin airport to destination airport) has the most late arrivals?

```python
df_flights_cleaned
```

```python
df_flights_cleaned[df_flights_cleaned['OriginAirportName'] == df_flights_cleaned['DestAirportName']]
```

```python
df_flights_grouped = df_flights_cleaned[df_flights_cleaned['ArrDel15'] == True]
df_flights_grouped['Airport Route'] = df_flights_cleaned['OriginAirportName'] + " - " + df_flights_cleaned['DestAirportName']
df_flights_grouped = df_flights_grouped.groupby('Airport Route')['ArrDel15'].count()
df_flights_grouped = df_flights_grouped.reset_index()
df_flights_grouped = df_flights_grouped.sort_values('ArrDel15', ascending=False)
df_flights_grouped.index = range(df_flights_grouped.shape[0])
df_flights_grouped.index = df_flights_grouped.index + 1
df_flights_grouped.rename({'ArrDel15': 'Arrivals'}, axis=1, inplace=True)
df_flights_grouped
```

```python
print('The route (from origin airport to destination airport) that has the most late arrivals is: \n{}, which presents {} late arrivals'.format(df_flights_grouped['Airport Route'][1], df_flights_grouped['Arrivals'][1]))
```

#### Which route has the highest average arrival delay?

```python
df_flights_grouped = df_flights_cleaned[df_flights_cleaned['ArrDelay'] > 0]
df_flights_grouped['Airport Route'] = df_flights_cleaned['OriginAirportName'] + " - " + df_flights_cleaned['DestAirportName']
df_flights_grouped = df_flights_grouped.groupby('Airport Route')['ArrDelay'].mean()
df_flights_grouped = df_flights_grouped.reset_index()
df_flights_grouped = df_flights_grouped.sort_values('ArrDelay', ascending=False)
df_flights_grouped.index = range(df_flights_grouped.shape[0])
df_flights_grouped.index = df_flights_grouped.index + 1
df_flights_grouped.rename({'ArrDelay': 'AvgArrDelay'}, axis=1, inplace=True)
df_flights_grouped
```

```python
print('The route (from origin airport to destination airport) that has the highest average arrival delay is: \n{}, which presents an average delay of {} minutes'.format(df_flights_grouped['Airport Route'][1], df_flights_grouped['AvgArrDelay'][1]))
```
